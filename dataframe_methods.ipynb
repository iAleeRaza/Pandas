```python
import pandas as pd
import numpy as np
```

```python
ipl = pd.read_csv('data\\ipl-matches.csv')
ipl.head(2)
```

```python
# use of value counts
# Very useful for series (dataframe columns)
# task: find which player has won the most POTM in finals, qualifiers or eliminators
ipl[~ipl['MatchNumber'].str.isdigit()]['Player_of_Match'].value_counts()
```

```python
# toss decision plot
ipl['TossDecision'].value_counts().plot(kind = 'pie')
```

```python
# how many matches each team has played
(ipl['Team1'].value_counts().sort_index() + ipl['Team2'].value_counts().sort_index()).sort_values(ascending = False)
```

```python
movies = pd.read_csv('data\\movies.csv')
movies.head(2)
```

```python
# sort_values
movies.sort_values('title_x')
```

```python
# sort_values with NaN values
# default temporary (inplace = False)
record = pd.DataFrame({
    'Name': ['Alice', 'Bob', np.nan, 'Charlie', 'Dave', 'Ali'],
    'Roll No:': ['K11', np.nan ,'K12', 'K13', 'K14', np.nan],
    'Age': [np.nan, 25, 30, np.nan, 35, 40],
    'Score': [90, 80, np.nan, 70, np.nan, np.nan]

})
record
```

```python
record.sort_values('Name') #by default send the NaN values the name column to the last
```

```python
record.sort_values('Name', na_position = 'first') 
```

```python
# sorting with multiple columns
movies.sort_values(['year_of_release', 'title_x'])
# will sort values 1st with YOR then by title_x
```

```python
movies.sort_values(['year_of_release', 'title_x'], ascending=[True, False])
```

```python
# rank (series)
batsman = pd.read_csv('data\\batsman_runs_ipl.csv')
batsman.head()
```

```python
batsman['batsman_rank'] = batsman['batsman_run'].rank(ascending = False)
batsman.sort_values('batsman_rank')
```

```python
# sort index
movies.sort_index(ascending = False)
```

```python
#set_index (default: inplace = False)
batsman.set_index('batter', inplace = True)
batsman
```

```python
#reset index
batsman.reset_index() #not permanant
```

```python
# making batsman_rank as index

batsman.reset_index().set_index('batsman_rank')
```
